```python
import re

def normalize_arabic(text):
    text = re.sub(r'[\u064B-\u065F\u0670]', '', text)
    text = re.sub("[إأآا]", "ا", text)
    text = re.sub("ى", "ي", text)
    text = re.sub("ؤ", "ء", text)
    text = re.sub("ئ", "ء", text)
    text = re.sub("ة", "ه", text)
    text = re.sub("گ", "ك", text)
    text = re.sub("ڤ", "ف", text)
    text = re.sub("چ", "ج", text)
    text = re.sub("پ", "ب", text)
    text = re.sub("ڜ", "ش", text)
    text = re.sub("ڪ", "ك", text)
    text = re.sub("ڧ", "ق", text)
    text = re.sub("ٱ", "ا", text)
    return text
```

```python
import emoji

def handle_emojis(text, mode='remove'):
    if mode == 'remove':
        return emoji.replace_emoji(text, '')
    elif mode == 'description':
        return emoji.demojize(text, language='ar')
    return text
```

```python
def handle_numbers_and_special_chars(text, mode='remove'):
    if mode == 'remove':
        text = re.sub(r'\d+', '', text)
        return text
    elif mode == 'normalize':
        number_map = {
                      '٠': 'صفر',
                      '١': 'واحد',
                      '٢': 'اثنان',
                      '٣': 'ثلاثة',
                      '٤': 'أربعة',
                      '٥': 'خمسة',
                      '٦': 'ستة',
                      '٧': 'سبعة',
                      '٨': 'ثمانية',
                      '٩': 'تسعة',
                      '0': 'صفر',
                      '1': 'واحد',
                      '2': 'اثنان',
                      '3': 'ثلاثة',
                      '4': 'أربعة',
                      '5': 'خمسة',
                      '6': 'ستة',
                      '7': 'سبعة',
                      '8': 'ثمانية',
                      '9': 'تسعة'
                  }
        for arabic, name in number_map.items():
            text = text.replace(arabic, name)
        return text
```

```python
arabic_punctuations = '''`÷×؛<>_()*&^%][ـ،/:"؟.,'{}~¦+|!”…“–ـ'''
english_punctuations = r"""!"#$%&'()*+,-./:;<=>?@[\]^_`{|}~"""

all_punctuations = arabic_punctuations + english_punctuations

def remove_arabic_punctuations(text):
    return re.sub(f"[{re.escape(all_punctuations)}]", " ", text)
```

```python
df['cleaned_text']=df['cleaned_text'].apply(remove_arabic_punctuations)
```

```python
def remove_diacritics(text):
    arabic_diacritics = re.compile("""
         ّ    |  # Shadda
         َ    |  # Fatha
         ً    |  # Tanwin Fath
         ُ    |  # Damma
         ٌ    |  # Tanwin Damm
         ِ    |  # Kasra
         ٍ    |  # Tanwin Kasr
         ْ    |  # Sukun
         ـ       # Tatwil (Kashida)
    """, re.VERBOSE)

    return re.sub(arabic_diacritics, '', text)
```

```python
df['cleaned_text']=df['cleaned_text'].apply(remove_diacritics)
```

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem.isri import ISRIStemmer
nltk.download('punkt_tab')
nltk.download('punkt')
nltk.download('stopwords')

def preprocess_arabic_text(text):

    tokens = word_tokenize(text)

    arabic_stopwords = set(stopwords.words('arabic'))

    stemmer = ISRIStemmer()

    processed = [stemmer.stem(word) for word in tokens if word not in arabic_stopwords]

    return " ".join(processed)
```

```python
df['cleaned_text']=df['cleaned_text'].apply(preprocess_arabic_text)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Vectorization using TF-IDF
vectorizer = TfidfVectorizer()
vec_data = vectorizer.fit_transform(df['cleaned_text'])

def search_documents(query, top_similar=10):
    # Preprocess the query
    query_preprocessed = normalize_arabic(query)
    query_preprocessed = handle_emojis(query_preprocessed)
    query_preprocessed = handle_numbers_and_special_chars(query_preprocessed)
    query_preprocessed = remove_arabic_punctuations(query_preprocessed)
    query_preprocessed = remove_diacritics(query_preprocessed)
    query_preprocessed = preprocess_arabic_text(query_preprocessed)

    # Vectorize the query
    query_vec = vectorizer.transform([query_preprocessed])

    # Calculate Cosine similarity
    similarities = cosine_similarity(query_vec, vec_data).flatten()

    # Sort the result and Retrieve the most 10 documents similar to the Query
    top_k_indices = similarities.argsort()[::-1][:top_similar]

    return df.iloc[top_k_indices]
```
